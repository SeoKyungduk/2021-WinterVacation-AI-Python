# 11_pandas

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(np.arange(1, 13).reshape(3, 4), columns = ['a', 'b', 'c', 'd']) # 3행 4열
df
```

```python
# 실습하기. df의 0행 a열의 데이터를 100으로 수정해주세요
df.loc[0, 'a'] = 100
df
```

```python
# 실습하기. 0행의 a, b 열을 100, 101로 수정해주세요
df.loc[0, ['a', 'b']] = [100, 101]
df
```

```python
# df의 3행에 13, 14, 15, 16
# - 수동으로 행을 추가하기
df.loc[3, :] = np.arange(13, 17)
df
```

```python
pd.DataFrame([[17, 18, 19, 20]], columns = ['a', 'b', 'c', 'd'])
```

```python
# 함수를 이용하여 행을 추가하기
new_df = pd.DataFrame([[17, 18, 19, 20]], columns = ['a', 'b', 'c', 'd'])
df = df.append(new_df)
```

```python
df.append(new_df)
```

```python
# index 초기화하기
df = df.reset_index(drop=True)
```

```python
# 삭제하기
# - 행 삭제
df.drop([0, 3]) # 0, 3행 삭제
```

```python
# - 열 삭제
df.drop('a', axis=1) # a열 삭제
```

```python
# 실습하기. df의 b와 c열을 삭제한 결과를 보여주세요.
df.drop(['b', 'c'], axis = 1)
```

```python
df.loc[:, ['a', 'b']]
```

```python
# 실습하기. df의 e열을 추가하면서 one, twom three, one, two의 값을 넣어주세요
df.loc[:, 'e'] = ['one', 'two', 'three', 'one', 'two']
df
```

```python
# df의 e열의 데이터가 one인 데이터만 조회해주세요
df.loc[df.e == 'one', :]
```

```python
# df의 e열의 데이터가 one 혹은 two인 데이터만 조회하기
df.e.isin(['one','two'])

df.loc[df.e.isin(['one','two']), :]
```

```python
# 실습하기. f열을 추가하면서 df의 e열의 데이터가 one 혹은 three인 행에만 데이터를 1로 넣어주세요
df.e.isin(['one', 'three'])
df.loc[df.e.isin(['one', 'three']), :]
df.loc[df.e.isin(['one', 'three']), 'f'] = 1
df
```

```python
# 결측치 제외하고 조회하기
df.dropna()
df.dropna(how='any')
df.dropna(how='all')
```

```python
# 결측치가 존재하는지 확인하기
df.isna()
df.isnull()
```

```python
# 실습하기. 결측치가 몇개 존재하는지 컬럼별로 확인해주세요
# - hint: 컬럼별 연산
# - hint: 논리값은 0과 1이라는 숫자
df.isna().sum()
```

```python
# 전체를 대상으로 하는 경우
df.isna().sum().sum()
```

```python
# 결축치의 값 채우기
df.fillna(5)
```
