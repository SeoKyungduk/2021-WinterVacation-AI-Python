### **bar 플랏**

```python
import seaborn as sns
iris = sns.load_dataset('iris')
```

```python
iris.sepal_length[:20].plot(kind = 'bar')
```

```python
df2 = iris.groupby(iris.species).mean()
df2.plot(kind = 'bar')
plt.xlabel('평균')
plt.ylabel('종')
```

### **Pie 차트**

```python
df3 = iris.species.value_counts()
df3.plot(kind = 'pie', autopct = '%.2f%%') # 소수점 2자리까지 나타냄
```

### **seaborn을 활용한 시각화하기**

```python
tips = sns.load_dataset('tips')
```

```python
# 총 주문 금액 대비 팁
sns.relplot(data = tips, x = tips.total_bill, y = tips.tip)
```

```python
# 여자와 남자의 팁 금액
sns.relplot(x = tips.sex, y = tips.tip, data = tips)
```

```python
# 남여의 비율보기
sns.barplot(data = tips, x = tips.sex, y = tips.tip)
```

```python
# 카테고리형 자료에 대해서 시각화하기
sns.relplot(data = tips, x = tips.total_bill, y = tips.tip, hue = tips.smoker)
```

```python
sns.relplot(data = tips, x = 'total_bill', y = 'tip', hue = 'smoker', size = 'size')
```

```python
sns.relplot(data = tips, x = 'total_bill', y = 'tip', hue = 'size', size = 'size')
```

```python
# 요일을 분석하기
tips.day.value_counts()
```

```python
# 점심과 저녁으로 카테고리를 나눠서 시각화하기
sns.catplot(data = tips, x = 'time', y = 'tip')
```

```python
# 실습하기. 점심과 저녁으로 카테고리를 나눠서 시각화하기
# - 흡연자의 정보를 추가해주세요.
sns.catplot(data = tips, x = 'time', y = 'tip', hue = 'smoker')
```

```python
# 실습하기. 요일별 catplot을 그려주세요.
sns.catplot(data = tips, x = 'day', y = 'tip')
```

```python
sns.swarmplot(data = tips, x = 'day', y = 'tip')
```
